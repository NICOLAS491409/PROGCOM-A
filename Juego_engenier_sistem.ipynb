```python
import re
import os

# ============================================================
# Funciones auxiliares
# ============================================================

def limpiar_pantalla():
    os.system("cls" if os.name == "nt" else "clear")

def pedir_entero(msg, min_val, max_val):
    while True:
        try:
            n = int(input(msg))
            if min_val <= n <= max_val:
                return n
        except:
            pass
        print(f"Ingresa un n√∫mero entre {min_val} y {max_val}.")

# ============================================================
# Funci√≥n para agregar reglas personalizadas del equipo
# ============================================================

def crear_reglas(cantidad, reglas_previas):
    nuevas_reglas = reglas_previas.copy()
    print("\nCREACI√ìN DE REGLAS")
    print("------------------------")
    print(f"Debes crear {cantidad} reglas nuevas.")
    print("Las reglas deben escribirse como texto, NO como c√≥digo.\n")

    for i in range(cantidad):
        descripcion = input(f"Regla #{len(nuevas_reglas)+1} (texto): ")

        # Creamos una regla *v√°lida pero gen√©rica*
        # El profesor despu√©s puede editarla
        nuevas_reglas.append(
            (descripcion, lambda p: True)
        )

    return nuevas_reglas

# ============================================================
# Funci√≥n para validar contrase√±a contra reglas
# ============================================================

def validar_contrasena(password, reglas):
    print("\nRESULTADOS:")
    print("-----------------------------------")
    exito = True

    for i, (descripcion, regla) in enumerate(reglas, start=1):
        try:
            cumple = regla(password)
        except:
            cumple = False

        if cumple:
            print(f"‚úî Regla {i}: CUMPLE ‚Äì {descripcion}")
        else:
            print(f"‚úò Regla {i}: NO CUMPLE ‚Äì {descripcion}")
            exito = False

    print("-----------------------------------")
    return exito

# ============================================================
# Reglas base programadas (primer equipo)
# ============================================================

reglas_base = [
    ("La contrase√±a debe tener m√≠nimo 10 caracteres.",
     lambda p: len(p) >= 10),

    ("Debe incluir al menos una may√∫scula.",
     lambda p: any(c.isupper() for c in p)),

    ("Debe incluir al menos un n√∫mero.",
     lambda p: any(c.isdigit() for c in p)),

    ("Debe incluir al menos un s√≠mbolo.",
     lambda p: any(not c.isalnum() for c in p)),

    ("No puede tener espacios.",
     lambda p: " " not in p)
]

# ============================================================
# JUEGO PRINCIPAL
# ============================================================

def jugar():
    limpiar_pantalla()
    print("=====================================")
    print(" JUEGO DE LA CONTRASE√ëA")
    print("=====================================")

    reglas_actuales = reglas_base.copy()
    equipos = int(input("¬øCu√°ntos equipos van a jugar? "))

    for equipo in range(1, equipos + 1):

        limpiar_pantalla()
        print(f"\n===== EQUIPO {equipo} =====")

        if equipo == 1:
            print("Este equipo usa las primeras 5 reglas base.")
        else:
            cantidad_nuevas = 5
            print(f"Debes agregar {cantidad_nuevas} reglas nuevas.")
            reglas_actuales = crear_reglas(cantidad_nuevas, reglas_actuales)

        input("\nPresiona ENTER para continuar al reto...\n")

        limpiar_pantalla()
        print(f"=== EQUIPO {equipo}: CREA TU CONTRASE√ëA ===")
        print("Debe cumplir TODAS las reglas actuales.")
        print(f"Total de reglas: {len(reglas_actuales)}\n")

        password = input("Ingresa tu contrase√±a: ")

        limpiar_pantalla()
        print(f"=== Validando contrase√±a del Equipo {equipo} ===")

        if validar_contrasena(password, reglas_actuales):
            print("\nüéâ ¬°El equipo ha cumplido TODAS las reglas!")
        else:
            print("\n‚ùå El equipo NO ha cumplido todas las reglas.")

        input("\nPresiona ENTER para continuar al siguiente equipo...\n")

    print("\n\n=== FIN DEL JUEGO ===")
    print("Gracias por jugar Gamer Quest Three.")


# Ejecutar el juego
if __name__ == "__main__":
    jugar()
```
