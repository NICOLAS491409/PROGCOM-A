PRIMER EJERCICIO

```python
import json

# 1. Crear una lista vacía para guardar las mascotas
mascotas = []

# 2. Pedir datos de 2 mascotas
for i in range(2):
    print(f"Ingresa los datos de la mascota {i+1}:")
    nombre = input("Nombre: ")
    tipo = input("Tipo (perro, gato, etc.): ")
    edad = int(input("Edad: "))

    # Crear diccionario para cada mascota
    mascota = {
        "nombre": nombre,
        "tipo": tipo,
        "edad": edad
    }

    mascotas.append(mascota)
    print()

# 3. Crear estructura final para JSON
datos_json = {
    "mascotas": mascotas
}

# 4. Guardar en un archivo JSON
with open("mascotas.json", "w", encoding="utf-8") as archivo:
    json.dump(datos_json, archivo, indent=4)

print("Archivo 'mascotas.json' creado correctamente.\n")

# 5. Leer el archivo JSON y mostrar los datos
with open("mascotas.json", "r", encoding="utf-8") as archivo:
    contenido = json.load(archivo)

print("Contenido del archivo:")
for m in contenido["mascotas"]:
    print(f"- {m['nombre']} ({m['tipo']}), {m['edad']} años")
```

SEGUNDO EJERCICIO

```python
import json

# 1. Lista vacía para guardar tareas
tareas = []

# 2. Pedir datos de 3 tareas al usuario
for i in range(3):
    print(f"Tarea {i+1}:")
    descripcion = input("Descripción: ")
    estado = input("Estado (pendiente / completada): ").lower()

    # Crear diccionario para la tarea
    tarea = {
        "descripcion": descripcion,
        "estado": estado
    }

    tareas.append(tarea)
    print()

# 3. Crear estructura final para JSON
datos_json = {
    "tareas": tareas
}

# 4. Guardar en archivo JSON
with open("tareas.json", "w", encoding="utf-8") as archivo:
    json.dump(datos_json, archivo, indent=4)

print("Archivo 'tareas.json' creado correctamente.\n")

# 5. Leer archivo y mostrar tareas pendientes
with open("tareas.json", "r", encoding="utf-8") as archivo:
    contenido = json.load(archivo)

print("Tareas pendientes:")
for tarea in contenido["tareas"]:
    if tarea["estado"] == "pendiente":
        print(f"- {tarea['descripcion']}")
```
